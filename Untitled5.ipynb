```python
doc = fitz.open("LenderFeesWorksheetNew (1).pdf")
page = doc[0]
page_dict = page.get_text("dict")

structured_data = {}
counter = 1

for block in page_dict.get("blocks", []):
    if "lines" not in block:
        continue
    for line in block["lines"]:
        for span in line.get("spans", []):
            key = f"text{counter}"
            structured_data[key] = {"text": span["text"], "bbox": span["bbox"]}
            counter += 1

json_output = json.dumps(structured_data, indent=2)
print(json_output)
```
